```python
import pandas as pd
import numpy as np

# Sample datetime columns (replace with your own column names if needed)
# Ensure Transaction_Date is in datetime format
df['Transaction_Date'] = pd.to_datetime(df['Transaction_Date'])

# If 'Transaction_Time' is a string or int (e.g., 1330), convert to HH:MM:SS
df['Transaction_Hour'] = df['Transaction_Time'].astype(str).str.zfill(4).str[:2].astype(int)

# 1. Hour of the day
df['HourOfDay'] = df['Transaction_Hour']

# 2. Night flag (12 a.m. to 6 a.m.)
df['Is_Night'] = df['HourOfDay'].apply(lambda x: 1 if 0 <= x <= 6 else 0)

# 3. Weekend flag
df['Is_Weekend'] = df['Transaction_Date'].dt.weekday.apply(lambda x: 1 if x >= 5 else 0)

df
```

```python
# Money based key features 
import pandas as pd
import numpy as np

# 1. Balance_After_Transaction
df['Balance_After_Transaction'] = df['Account_Balance'] - df['Transaction_Amount']

# 2. Amount_Per_Balance
# Avoid division by zero by replacing 0 or near-zero balances with NaN
df['Amount_Per_Balance'] = df.apply(
    lambda row: row['Transaction_Amount'] / row['Account_Balance'] if row['Account_Balance'] > 0 else 0,
    axis=1
)

# 3. Is_Large_Transaction (flag transactions above 95th percentile)
threshold = df['Transaction_Amount'].quantile(0.95)
df['Is_Large_Transaction'] = (df['Transaction_Amount'] > threshold).astype(int)

# 4. Daily_Spend
# Ensure Transaction_Date is in datetime format
df['Transaction_Date'] = pd.to_datetime(df['Transaction_Date'])

# Assumes you have 'Customer_ID' to group by â€” replace with appropriate identifier if different
if 'Customer_ID' in df.columns:
    df['Daily_Spend'] = df.groupby(['Customer_ID', df['Transaction_Date'].dt.date])['Transaction_Amount'].transform('sum')
else:
    print('Customer_ID column not found; skipping Daily_Spend. Provide user/account identifier.')
# Final check
print(df[['Account_Balance', 'Transaction_Amount', 'Balance_After_Transaction', 
          'Amount_Per_Balance', 'Is_Large_Transaction']].head())
```

```python
# Extract hour
df['Transaction_Hour'] = pd.to_datetime(df['Transaction_Time'], format='%H:%M:%S').dt.hour

# Define anomaly rules (you can tune these)
def detect_device_hour_anomaly(row):
    if row['Transaction_Device'] == 'POS' and (row['Transaction_Hour'] < 5 or row['Transaction_Hour'] > 23):
        return 1
    if row['Transaction_Device'] == 'ATM' and row['Transaction_Hour'] < 4:
        return 1
    return 0

df['Device_Hour_Anomaly'] = df.apply(detect_device_hour_anomaly, axis=1)
```

```python
# Assume the user's most frequent city is their 'home'
user_home_city = df.groupby('Customer_ID')['City'].agg(lambda x: x.mode().iloc[0])

df['User_Home_City'] = df['Customer_ID'].map(user_home_city)
#df['User_Home_State'] = df.groupby('Customer_ID')['State'].agg(lambda x: x.mode().iloc[0])

# Flag mismatch
df['Device_Location_Mismatch'] = ((df['Transaction_Location'] != df['User_Home_City']) | 
                                   (df['City'] != df['User_Home_City'])).astype(int)
df.to_csv('C:/Users/kau75421/LLMprojects/Fraudetector/bank_model_project/data/processed/Bank_Transaction_Fraud_Detection_Processed.csv', index=False)
```
